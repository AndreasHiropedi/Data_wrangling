We will then clean up the data. First, we store the data in a list, and then itterate over the list of dataframes to clean up each individual data. For the cleaning of the data, we will remove any missing values, and removing the unnecesary columns (only keep the columns neeeded for calculations).

```python
# store all the dataframes in one list
data = [aberporth_data, armagh_data, ballypatrick_data, bradford_data, braemar_data, camborne_data, cambridge_data,
        cardiff_data, chivenor_data, cwmystwyth_data, dunstaffnage_data, durham_data, eastbourne_data, 
        eskdalemuir_data, heathrow_data, hurn_data, lerwick_data, leuchars_data, lowestoft_data, manston_data, 
        nairn_data, newtonrigg_data, oxford_data, paisley_data, ringway_data, rossonwye_data, shawbury_data, 
        sheffield_data, southampton_data, stornoway_data, suttonbonington_data, tiree_data, valley_data, 
        waddington_data, whitby_data, wickairport_data, yeovilton_data]
# cleaning up the data
for data_item in data:
        data_item['sun'] = pd.to_numeric(data_item['sun'].map(lambda word: str(word).rstrip('#*$')), errors = 'coerce')
        data_item['rain'] = pd.to_numeric(data_item['rain'].map(lambda word: str(word).rstrip('#*$')), errors = 'coerce')
```

After cleaning all the data, we will concatenate all the dataframes.

```python
# merge all the data, and sort it in ascending order
all_data = pd.concat(data)
all_data = all_data.sort_values(by = ['yyyy'])
```

Now, after concatenating all the data, we will group the data by decades, and then we will compute the mean rainfall, as required, in mm/month for the decades 1850s and 1990s, and round it to 2 decimal places.

```python
# group data by decades and drop the sunlight data
no_sun = all_data.drop('sun', axis = 1)
all_data_by_decade = no_sun.groupby(pd.cut(no_sun['yyyy'], np.arange(1849, 2000, 10)))
# compute the mean rainfall for each decade
# then compute the mean of all decade means
rainfall = all_data_by_decade.mean().round(2)
mean_rainfall_1850s = rainfall['rain'].iloc[0]
mean_rainfall_1990s = rainfall['rain'].iloc[len(rainfall) - 1]
all_data_by_decade = no_sun.groupby(pd.cut(no_sun['yyyy'], np.arange(1849, 2020, 10)))
rainfall = all_data_by_decade.mean().round(2)
```

Now, after sorting all the data, we will compute the mean rainfall, as required, in mm/month for the decades 1850s and 1990s, and round it to 2 decimal places.

```python
# group the data by year (one value per year) and drop the rainfall data
no_rain = all_data.drop('rain', axis = 1)
all_data_by_year = no_rain.groupby(pd.cut(no_rain['yyyy'], np.arange(1898, 2000, 1)))
# compute the mean sunlight and drop the year column
# then compute the mean of all they yearly means
sunlight = all_data_by_year.mean().round(2)
mean_sunlight_1899 = sunlight['sun'].iloc[0]
mean_sunlight_1999 = sunlight['sun'].iloc[len(sunlight) - 1]
all_data_by_year = no_rain.groupby(pd.cut(no_rain['yyyy'], np.arange(1898, 2021, 1)))
sunlight = all_data_by_year.mean().round(2)
```

Lastly, we will compute the rows used in the calculations. This will consist of five parts: first, we will calulate the number of rows used to compute the mean rainfall for the 1850s; next, we will calulate the number of rows used to compute the mean rainfall for the 1990s; then, we will calulate the number of rows used to compute the mean sunlight for 1899; finally, we will calulate the number of rows used to compute the mean sunlight for 1999. The last part will simply be adding all these lines together.

```python
# first, compute the number of rows used for 1850s mean rainfall
mean_rainfall_1850s_rows = len((all_data[(all_data['yyyy'] >= 1850) & (all_data['yyyy'] < 1860) & (all_data['rain'].notnull())]))

# next, compute the number of rows used for 1990s mean rainfall
mean_rainfall_1990s_rows = len((all_data[(all_data['yyyy'] >= 1990) & (all_data['yyyy'] < 2000) & (all_data['rain'].notnull())]))

# now, compute the number of rows used for 1899 mean sunlight
mean_sunlight_1899_rows = len((all_data[(all_data['yyyy'] == 1899) & (all_data['sun'].notnull())]))

# lastly, compute the number of rows used for 1999 mean sunlight
mean_sunlight_1999_rows = len((all_data[(all_data['yyyy'] == 1999) & (all_data['sun'].notnull())]))

# now, compute the total number of rows
total_rows = mean_rainfall_1850s_rows + mean_rainfall_1990s_rows + mean_sunlight_1899_rows + mean_sunlight_1999_rows
```

Finally, we will display every data in a nice manner, such as the means and number of rows being used. This is so that it is easy to read.

```python
print("Mean rainfall in mm/month measured at UK weather stations for the 1850s:", mean_rainfall_1850s)
print("Mean rainfall in mm/month measured at UK weather stations for the 1990s:", mean_rainfall_1990s)
print("Mean sunlight in hours/month measured at UK weather stations for the year 1899:", mean_sunlight_1899)
print("Mean sunlight in hours/month measured at UK weather stations for the year 1999:", mean_sunlight_1999)
print("The number of rows used to compute the mean rainfall and mean sunlight:", total_rows)
```
